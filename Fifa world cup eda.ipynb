### visualisation

```python
import seaborn as sns
```

```python
fifa['possession team1'] = fifa['possession team1'].replace('%', '', regex=True)
fifa['possession team1'] = pd.to_numeric(fifa['possession team1'], errors='coerce', downcast='float')
fifa['possession team2'] = fifa['possession team2'].replace('%', '', regex=True)
fifa['possession team2'] = pd.to_numeric(fifa['possession team2'], errors='coerce', downcast='float')
```

```python
team1_possession = fifa["possession team1"]
team2_possession = fifa["possession team2"]

plt.bar(["team1", "team2"], [team1_possession.mean(), team2_possession.mean()])
plt.xlabel("Team")
plt.ylabel("Average Possession Percentage")
plt.show()
```

```python
plt.scatter(fifa["possession in contest"],fifa["number of goals team1"],label="Team 1")
plt.scatter(fifa["possession in contest"],fifa["number of goals team2"],label="Team 2")
plt.xlabel("possession in contest")
plt.ylabel("number of goals by team 1 and team 2")
plt.legend()
plt.show()
```

```python
team1_total_attempts = fifa["total attempts team1"]
team2_total_attempts = fifa["total attempts team2"]
team1_on_target = fifa["on target attempts team1"]
team2_on_target = fifa["on target attempts team2"]
team1_penalty_attempts = fifa["attempts inside the penalty area team1"]
team2_penalty_attempts = fifa["attempts inside the penalty area  team2"]

fig, axs = plt.subplots(1, 3, figsize=(15, 5))
axs[0].bar(["team1", "team2"], [team1_total_attempts.mean(), team2_total_attempts.mean()])
axs[0].set_title("Total Attempts")
axs[1].bar(["team1", "team2"], [team1_on_target.mean(), team2_on_target.mean()])
axs[1].set_title("On Target Attempts")
axs[2].bar(["team1", "team2"], [team1_penalty_attempts.mean(), team2_penalty_attempts.mean()])
axs[2].set_title("Attempts Inside Penalty Area")
plt.show()
```
